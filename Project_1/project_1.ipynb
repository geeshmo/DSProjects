## Step 4: Visualize the data

##### 14. Using MatPlotLib and PyPlot, plot the distribution of the Rate using histograms.

```python
import seaborn as sns
%matplotlib inline
def distribution_plotter(column, data):
    data = data[column]
    sns.set(rc={"figure.figsize": (6, 3)})
    sns.set_style("dark")
    dist = sns.distplot(data, hist_kws={'alpha':0.2}, kde_kws={'linewidth':2, 'color':'blue'}, color ='red')
    dist.set_title('Distribution of ' + column + '\n', fontsize=12)

distribution_plotter('Rate', satdatadict2)

```
