# Graphing the Following Equations and Creating Contur Plots

1. $F_1(x) = Exp(-0.5*(x-2)^2)$
2. $F_2(x) = x^2 * Exp(-0.8*(x+2)^2)$
3. $F_3(x) = F_2(x)*F_1(x)$
4. $F_4(x) = \frac{1}{1+Exp(-x)}$
5. $G_1(x,y) = Exp(-0.01*(x^3+y^3))$
6. $G_2(x,y) = Log(x^2+2*y^2)*x^2$

```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

X = np.arange(-10,11,0.01)

F = lambda x : np.exp(-0.5 * (x - 2)**2)

G = lambda x : x**2 * np.exp(-0.8 * (x + 2)**2)

H = lambda x : G(x) * F(x)

P = lambda x : 1/(1+ np.exp(-x))


fig = plt.figure(figsize=(8, 8))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
ax1.plot(F(X), label = 'exp(-0.5 *(x-2)**2)')
ax2.plot(G(X), label = 'x**2 *exp(-0.8 *(x+2)**2)')
ax3.plot(H(X), label = 'F(x)*G(x)')
ax4.plot(P(X), label = '1(1+exp(-x))')


ax1.legend(loc ='upper left', fontsize ='small')
ax2.legend(loc ='upper left', fontsize ='small')
ax3.legend(loc ='upper left', fontsize ='small')
ax4.legend(loc ='upper left', fontsize ='small')
plt.show()
% matplotlib inline                
```

```python
from mpl_toolkits.mplot3d import Axes3D
X = np.linspace(-2,2, 100)
Y = np.linspace(-2,2, 100)
X, Y = np.meshgrid(X,Y)
Z = np.exp(-X**2 - Y**2)
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,2,1, projection = '3d')
ax1.plot_surface(X, Y, Z)
ax2 = fig.add_subplot(1,2,2)
ax2.contour(X, Y, Z)
```

```python
#G1(x,y)=Exp(−0.01∗(x3+y3))
 
#G2(x,y)=Log(x2+2∗y2)∗x2


from mpl_toolkits.mplot3d import Axes3D
X = np.linspace(-2,2, 100)
Y = np.linspace(-2,2, 100)
X, Y = np.meshgrid(X,Y)
Z = np.exp(-0.01*(X**3 +Y**3))
P = np.log(X**2 + 2*Y**2) * (X**2)
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(2,2,1, projection = '3d')
ax1.plot_surface(X, Y, Z)
ax2 = fig.add_subplot(2,2,2)
ax2.contour(X, Y, Z)
ax3 = fig.add_subplot(2,2,3, projection = '3d')
ax3.plot_surface(X, Y, P)
ax4 = fig.add_subplot(2,2,4)
ax4.contour(X, Y, P)
```
